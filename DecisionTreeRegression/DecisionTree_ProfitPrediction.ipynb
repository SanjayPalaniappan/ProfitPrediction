```python
# Importing required packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading the data present in CSV file
dataset = pd.read_csv('50_startups.csv')
dataset
```

```python
# Converting categorical columns into numerical value using One Hot Encoding
datasets = pd.get_dummies(dataset, drop_first=True)
datasets
```

```python
# Getting column names of dataset
datasets.columns
```

```python
# Assigning input and output into separate variables
independent = datasets[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = datasets[['Profit']]
```

```python
#Splitting into train and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
#Applying standardization to scale the data of all columns to bring into same scale
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#Building the model using standardized training data
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion = "absolute_error", splitter = "best", max_features = "auto")
regressor=regressor.fit(X_train,y_train)
```

```python
#Predicting built model on test data
y_pred=regressor.predict(X_test)
y_pred
```

```python
#Calculating R^2 score
from sklearn.metrics import r2_score
rSquared_value=r2_score(y_test,y_pred)
rSquared_value
```

```python
# Returns the importance of features/columns, in terms of values calculated
regressor.feature_importances_
```

```python
#Identifying the best split
regressor.max_features_
```

```python
# Decision tree Formation
regressor.tree_
```

```python
#Plotting and displaying tree
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```
